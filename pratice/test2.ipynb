```python
import yfinance as yf
import time

ticker = yf.Ticker("AAPL")

while True:
    data = ticker.history(period="1d", interval="1m")
    latest_price = data["Close"].iloc[-1]

    print("Live Price:", latest_price)

    time.sleep(5)
```
